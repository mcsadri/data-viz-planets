```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

#### Load seaborn's built-in planets dataset

```python
sns.get_dataset_names()
df = sns.load_dataset("planets")
display(df.head())
display(df.describe())
```

#### Display a scatterplot comparing all objects where x-axies is the Orbital Period, and the y-axis is the Mass

```python
sns.scatterplot(data = df, x = "mass", y = "distance", hue = "method")
```

```python
sns.histplot(data = df, x = "mass", kde = True)
```

```python
sns.boxplot(data = df, x = "number", y = "orbital_period")
```

```python
sns.pairplot(data = df, hue = "method")
plt.show
```
